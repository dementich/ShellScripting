# Базовые утилиты Bash

Самыми базовыми утилитами, озвученными в конспектируемом руководстве к Bash, являются ls, cat, cp, mv rm, less. Их здесь описываем очень избирательно: очевидно, что командами mv и cat можно уничтожить имеющийся файл. Можно настроить систему так, чтобы она спрашивала подтверждения - для этого существует флажок -i. Если есть желание, можно даже заставить систему использовать интерактивные версии команд. Делается это через создание псевдонима в файле .bashrc текущего пользователя (для пользователя root это /root/.bashrc): `alias mv='mv -i'`.

## Отправка файла на печать

Здесь работает команда `lpr <file>`. Посмотреть задачи принтера можно командой `lpq`.

## Поиск в файле

Здесь нужна команда `grep`. В самом базовом варианте она выглядит следующим образом: `grep '<text>' <file>`, где `'<text>'` - это разыскиваемый нами текст, а `<file>` - это имя файла. Грепить не обязательно в одном файле - можно вместо имени файла задать маску (если интересуют абсолютно все файлы, то маска выглядит как `*`). Если же грепить хочется и в дочерних каталогах, то есть опция `-r`: `grep -r '<text>' *`.

## Просмотр файлов

Кроме команд cat и less здесь можно использовать head и tail. По-умолчанию они выводят 10 первых (или последних) строчек. Это число можно установить например вот так: `head -<lines> <file>` - где `<lines>` - нужное нам число строк.

Если есть желание отсортировать строки файла в алфавитном порядке, можно воспользоваться командой `sort`. Она не меняет содержимого файл, просто выводит данные в отсортированном виде.

Также есть команда для вывода только уникальных строк файла - uniq.

Сравнить содержимое файлов можно командой diff: `diff <file1> <file2>`. В данном случае сначала выведутся строки файла `<file1>`, которых нет в `<file2>`, а потом - строки `<file2>`, которых нет в `<file1>`.

Посмотреть формат файла можно командой `file`.

## Просто полезные утилиты

1. Посмотреть дату можно командой date=).  
2. Можно записать историю выполнения команд bash с результатом с помощью команды script. Завершается запись командой exit, результат сохраняется в файле typescript.  
3. Можно переформатировать текстовый файл из формата linux в dos посредством команды `unix2dos <file>`, а командой `dos2unix <file>` - наоборот.
