# Начало работы с Linux

## Создание пользователя

Здесь все просто. Добавить пользователя можно командой `adduser`. Поменять пароль можно командой `passwd`.
Зайти в систему от имени пользователя - `login <username>`. Выйти из системы - команда `logout` или `Ctrl+d`. 
Удалить пользователя - команда `deluser`. Посмотреть всех пользователей системы можно выведя на консоль содержимое файла `/etc/passwd`.

## Горячие клавиши

`Ctrl+a` - перейти на начало строки  
`Ctrl+e` - перейти на конец строки  
`Ctrl+u` - удалить текст от текущей позиции курсора до начала строки  
`Ctrl+k` - удалить текст от текущей позиции курсора до конца строки  
`Ctrl+_` - отменить последнее действие  

## Текущий процессор командной строки

Узнать текущий процессор командной строки можно командой `echo $0`.

## Выполнение команд

Повторить прошлую команду можно выполнить командой `!!`. Эта операция полезна, если при выполнении прошлой команды требовалось вводить пароль администратора - система в данном случае не потребует его снова. 

Последняя выполненная команда содержится в переменной `!$`. Это удобно использовать, когда мы сначала с помощью `cat` вывели текст какого-то файла, а теперь хотим отправить его на принтер - `lpr !$`.

Повторить прошлую команду с заменой одной подстроки в команде на другую можно командой `^old^new^`
